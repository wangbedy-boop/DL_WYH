```python
from mxnet import init, np, npx
from mxnet.gluon import nn

npx.set_np()

net = nn.Sequential()
net.add(nn.Dense(8, activation='relu'))
net.add(nn.Dense(1))
net.initialize()  

X = np.random.uniform(size=(2, 4))
net(X) 
```

```python
print(net[1].params)
```

```python
print(type(net[1].bias))
print(net[1].bias)
print(net[1].bias.data())
```

```python
net[1].weight.grad()
```

```python
print(net[0].collect_params())
print(net.collect_params())#一次性访问所有参数
```

```python
net.collect_params()['dense1_bias'].data()
```

```python
#从嵌套块收集参数
def block1():
    net = nn.Sequential()
    net.add(nn.Dense(32, activation='relu'))
    net.add(nn.Dense(16, activation='relu'))
    return net

def block2():
    net = nn.Sequential()
    for _ in range(4):
        # 在这里嵌套
        net.add(block1())
    return net

rgnet = nn.Sequential()
rgnet.add(block2())
rgnet.add(nn.Dense(10))
rgnet.initialize()
rgnet(X)
```

```python
print(rgnet.collect_params)
print(rgnet.collect_params())
```
